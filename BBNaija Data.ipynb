```
# Import library
import GetOldTweets3 as got
text_tweets = []
```

```
#@title QUERY PARAMETERS { vertical-output: true, display-mode: "both" }
query = "BBNaijialockdown2020" #@param {type:"string"}
startDate = "2020-07-01" #@param {type:"date"}
stopDate = "2020-08-21" #@param {type:"date"}
numberOfTweets =  4000#@param {type:"integer"}
name = ""
def get_tweets():
  tweetCriteria = got.manager.TweetCriteria().setQuerySearch(query)\
                                           .setSince(startDate)\
                                           .setUntil(stopDate)\
                                           .setMaxTweets(numberOfTweets)
  # List of object get stored in 'tweets' variable
  tweets = got.manager.TweetManager.getTweets(tweetCriteria)

  print(tweets[0].__dict__)
  for tweet in tweets:
    data = {
        'username': tweet.username, 
        'to': tweet.to, 
        'text': tweet.text, 
        'retweets': tweet.retweets, 
        'favorites': tweet.favorites, 
        'replies': tweet.replies, 
        'id': tweet.id, 
        'permalink': tweet.permalink, 
        'author_id': tweet.author_id, 
        'date': tweet.date, 
        'formatted_date': tweet.formatted_date, 
        'hashtags': tweet.hashtags, 
        'mentions': tweet.mentions, 
        'geo': tweet.geo, 
        'urls': tweet.urls
    }
    print(data)
    text_tweets.append(data) 
  import pandas as pd
  df = pd.DataFrame(text_tweets)
  df.to_csv(query + '.csv')
  print(query +'.csv created successfully.')
  print(len(text_tweets))
get_tweets()
```
